# #Prediction using Decision Tree Algorithm

```python
#PREPARING DECISION TREE MODEL FROM IRIS DATA
```

```python
#INCLUDING ALL LIBRARIES
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
print(df.head())
```

```python
#training the iris data by train split function
from sklearn.model_selection import train_test_split
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

Decision tree Formation

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

Visualizing  Decision Tree

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(model,filled=True)
```

# Testing Model 

```python

prediction=model.predict(x_test)
prediction
df1=pd.DataFrame({'predicted value':prediction,'actual value':y_test})
#df1['label']=df['predicted value'].replce(dict(enumarate(iris.target_names)))
df1.head()
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(prediction,y_test)
print(score)
```
